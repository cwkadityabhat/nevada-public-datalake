```python
import boto3
import pandas as pd
```

```python
s3 = boto3.client('s3', region_name="us-west-2")
ssm = boto3.client("ssm", region_name="us-west-2")
```

```python
dlz_bucket = ssm.get_parameter(Name="SYSTEM.S3BUCKET.DLZ", WithDecryption=False)["Parameter"]["Value"]
lz_bucket = ssm.get_parameter(Name="SYSTEM.S3BUCKET.LZ", WithDecryption=False)["Parameter"]["Value"]
prefix = 'bronzehealth/Immunization/' # prefix of Dataset you have access to
```

```python
output_df = pd.DataFrame()
s3_obj_list = s3.list_objects(Bucket = dlz_bucket, Prefix = prefix)['Contents']
print(s3_obj_list)
    
```

```python
for obj in s3_obj_list:
    data = s3.get_object(Bucket=dlz_bucket, Key=obj.get('Key'))
    s3_data = data['Body']
    df = pd.read_csv(s3_data, header=0)
    output_df = pd.concat([output_df, df], ignore_index=True)
```
